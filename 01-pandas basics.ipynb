```python
url = "https://www.theguardian.com/football/laligafootball/table"
# go to the webpage, find all of the tables and save them in a variable tables
# this read_html looks for tables specifically
tables = pd.read_html(url)
len(tables)
```

```python
# find the table we want and make it into a df
df = tables[0]
df.head(0)
```

```python
# How to save a new CSV!!
df.to_csv("teams.csv", index=False)
```

#  Turn list of Dictionaries into Dataframes!!!

```python
import requests

url = "http://api.weatherapi.com/v1/forecast.json?key=1deea713c0524e4abd621042220711&q=London&days=3&aqi=yes&alerts=no"
response = requests.get(url)
data = response.json()
```

```python
data['forecast']
```

```python
# Convert into CSV, but not easily navigable, we want something more "tablish"
pd.DataFrame(data['forecast']['forecastday'] )            
```

```python
# Convert into CSV 
pd.json_normalize(data['forecast']['forecastday'] ) 
```
